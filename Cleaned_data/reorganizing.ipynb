```python
import pandas as pd
import re

# 读文件
df = pd.read_csv('output/shipper_draft.csv/shipper_raw.csv')
df.head()

# 定义拆分函数
def split_address_phone(text):
    parts = re.split(r'\\r|\\n|\n|\r', str(text))
    parts = [p.strip() for p in parts if p.strip()]

    address, phone, country = '', '', ''

    if len(parts) >= 2:
        address = parts[1]
    if len(parts) >= 3:
        phone_candidate = parts[-1]
        if re.search(r'\d', phone_candidate):
            phone = phone_candidate
            if len(parts) >= 3:
                address = ' '.join(parts[1:-1])
        else:
            address = ' '.join(parts[1:])

    # 国家判断
    if re.search(r'ecuador', text, re.I):
        country = 'Ecuador'
    elif re.search(r'peru', text, re.I):
        country = 'Peru'
    elif re.search(r'colombia', text, re.I):
        country = 'Colombia'
    elif re.search(r'united states', text, re.I):
        country = 'United States'
    elif re.search(r'canada', text, re.I):
        country = 'Canada'

    return pd.Series([address, phone, country])

# 应用拆分到新列
df[['address', 'phone', 'country']] = df['shipper_name_&_address_&_phone'].apply(split_address_phone)

# 去重
df_clean = df.drop_duplicates(subset=['shipper']).reset_index(drop=True)

# 添加 shipper_id
df_clean.insert(0, 'shipper_id', df_clean.index + 1)

# 保存结果
df_clean.to_csv('output/shipper_clean.csv', index=False)

df_clean.head()
```

```python
spark_df_pro.createOrReplaceTempView("port_list")
re = spark.sql(
'''
SELECT
    a.port_id,
    a.name,
    a.country,
    b.code,
    b.air,
    b.ocean,
    b.rail,
    b.road,
    b.mail
FROM (
    SELECT
        ROW_NUMBER() OVER (ORDER BY destination_port) AS port_id,
        destination_port AS name,
        destination_country AS country
    FROM shipment_fact
    WHERE destination_port != 'NaN'
    GROUP BY destination_port, destination_country
) a
JOIN (
    SELECT
        name,
        code,
        air,
        ocean,
        rail,
        road,
        mail
    FROM port_list
) b
ON a.name = b.name
order by port_id
'''
)

re.show(truncate=False)
```

```python
from pyspark.sql.functions import col, when

desc_col = col("description")

spark_df_item = spark_df_item.withColumn(
    "item",
    when(desc_col == "Banano Cavendish Organico Fairtrade con Banda - 100",
         "Organic Fairtrade Cavendish Banana with Band - 100")
    .when(desc_col == "Banano Cavendish Organico Fairtrade 3lb con Banda",
         "Organic Fairtrade Cavendish Banana 3lb with Band")
    .when(desc_col == "Banano Cavendish Organico Fairtrade 3lb con Banda Costco",
         "Organic Fairtrade Cavendish Banana 3lb with Costco Band")
    .when(desc_col == "Banano Cavendish Convencional Fairtrade con etiqueta",
         "Conventional Fairtrade Cavendish Banana with Label")
    .when(desc_col == "Banano Cavendish Organico Fairtrade con banda",
         "Organic Fairtrade Cavendish Banana with Band")
    .otherwise(desc_col)   # fallback if not matched
)

```
