```python
import requests
import pandas as pd
url = "https://www.fruityvice.com/api/fruit/all"

r = requests.get(url)
print(f"Status Code: {r.status_code}")
```

```python
if r.ok:
    data = r.json()
else:
    raise Exception(f"API call failed: {r.status_code}\n{r.text}")
```

```python
df = pd.json_normalize(data)
df.head()
```

```python
df_simple = df[[
    'name',
    'family',
    'genus',
    'order',
    'nutritions.carbohydrates',
    'nutritions.protein',
    'nutritions.fat',
    'nutritions.calories',
    'nutritions.sugar'
]]

df_simple.head()
```
